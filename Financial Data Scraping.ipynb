Scraping Income Statement data from Yahoo Finance

```python
from lxml import html
from lxml import html
import requests
import yfinance as yf
import numpy as np
import pandas as pd

def scrape_table(url):
    page = requests.get(url)
    tree = html.fromstring(page.content)
    table = tree.xpath("//h1/text()")
    print(table)
    table_rows = tree.xpath("//div[contains(@class, 'D(tbr)')]")
    
# Ensure that some table rows are found; if none are found, then it's possible
# that Yahoo Finance has changed their page layout, or have detected
# that you're scraping the page.
    try:
        assert len(table_rows) > 0

    except AssertionError as msg:
        print(msg)
    
    parsed_rows = []
    
    for table_row in table_rows:
        parsed_row = []
        el = table_row.xpath("./div")
    
        none_count = 0
    
        for rs in el:
            try:
                (text,) = rs.xpath('.//span/text()[1]')
                parsed_row.append(text)
            except ValueError:
                parsed_row.append(np.NaN)
                none_count += 1

        if (none_count < 4):
            parsed_rows.append(parsed_row)
    
    df = pd.DataFrame(parsed_rows)
    df = df.set_index(0)
    df = df.dropna()
    df = df.transpose()
    df = df.replace('-', '0')
    indexNames = df[(df['Breakdown'] == 'ttm')].index
    df.drop(indexNames , inplace=True)
    df = df.fillna(0)
    
    numeric_columns = list(df.columns)[1::] 

    for column_name in numeric_columns:
        df[column_name] = df[column_name].str.replace(',', '') # Remove the thousands separator
        df[column_name] = df[column_name].astype(np.float64) # Convert the column to float64

    return df
```

```python
#symbols = ['AAPL', 'MFT.NZ']
financials = {}
lss = []
#create a dictionary of ticker names and their respective statements' urls

import xlrd 
import google
import pandas as pd    
loc = r"C:\Users\admin\uk_1.xlsx"
  
wb = xlrd.open_workbook(loc) 
sheet = wb.sheet_by_index(0) 

for i in range(sheet.nrows):
    company = sheet.cell_value(i, 0)
    symbols =sheet.cell_value(i, 1)
    #for symbol in symbols:
    financials[symbols] = ['https://finance.yahoo.com/quote/' + symbols + '/financials?p=' + symbols]
    lss.append(company)

data = pd.DataFrame()

for f, ls in zip(financials, lss):
    try:
        countries = yf.Ticker(f).info['country']
        names = yf.Ticker(f).info['shortName']
    except IndexError:
        countries = 'null'
        names = 'null'
    except ValueError:
        countries = 'null'
        names = 'null'
    except KeyError as ke:
        print('Key Not Found in Dictionary:', ke)
        continue
    except AssertionError:
        print('Assertion error')
    df_income_statement = scrape_table(financials[f][0])
    df_income_statement.insert(1, 'Symbol', f)   
    df_income_statement.insert(2, 'Country', countries)
    df_income_statement.insert(3, 'name of companies', ls)
    data = data.append(df_income_statement)    
    
```

```python
import pandas
from openpyxl import load_workbook

book = load_workbook('Yahoo-Finance-Scrape-2020-09-23.xlsx')
writer = pandas.ExcelWriter('Yahoo-Finance-Scrape-2020-09-23.xlsx', engine='openpyxl')
writer.book = book
writer.sheets = {ws.title: ws for ws in book.worksheets}

for sheetname in writer.sheets:
    data.to_excel(writer,sheet_name=sheetname, startrow=writer.sheets[sheetname].max_row, index = False,header= True, na_rep='NA')

writer.save()
```
